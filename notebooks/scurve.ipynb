```python
import numpy as np
import matplotlib.pyplot as plt

# Define logistic function
def logistic_function(x, L, k, x0):
    return L / (1 + np.exp(-k*(x-x0)))

# Generate some sample data
years = np.arange(0, 40, 1)  # Years from 0 to 39
L = 1000  # Maximum number of patents
k = 0.2  # Growth rate parameter
x0 = 20  # Midpoint of growth

# Calculate cumulative number of patents
cumulative_patents = logistic_function(years, L, k, x0)

# Plot the logistic curve
plt.figure(figsize=(10, 6))
plt.plot(years, cumulative_patents, color='blue')
plt.ylabel('Kümülatif patent sayısı')

# Hide the tick labels on both axes
plt.xticks([])
plt.yticks([])

plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define logistic function
def logistic_function(x, L, k, x0):
    return L / (1 + np.exp(-k*(x-x0)))

# Generate some sample data
years = np.arange(0, 40, 1)  # Years from 0 to 39
L = 100  # Maximum market share percentage
k = 0.2  # Growth rate parameter
x0 = 20  # Midpoint of growth

# Calculate cumulative market share
cumulative_market_share = logistic_function(years, L, k, x0)

# Calculate technology adoption (first derivative of logistic function)
adoption_rate = k * cumulative_market_share * (1 - cumulative_market_share / L)

# Normalize the adoption rate for better visualization
adoption_rate = adoption_rate / max(adoption_rate) * 55

# Plot the logistic curve (cumulative market share)
plt.figure(figsize=(10, 6))
plt.plot(years, cumulative_market_share, color='orange', label='S-eğrisi Lojistik Modeli')

# Plot the normal distribution curve (technology adoption rate)
plt.plot(years, adoption_rate, color='blue', label='Teknoloji Benimseme Modeli')

# Hide the tick labels on both axes
plt.xticks([])
plt.yticks([])

# Add labels
plt.ylabel('Kümülatif patent sayısı')
plt.legend()
plt.grid(True)
plt.show()
```
