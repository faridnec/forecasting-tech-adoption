### Synthetic Curve with Seasonality

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def logistic_function(t, k, m, b):
    """
    Logistic function for adoption rate modeling.

    Parameters:
    - t: Time (can be an array of time points)
    - k: Growth rate (steepness of the curve)
    - m: Maximum value (asymptote)
    - b: Time shift (horizontal shift)

    Returns:
    - Values of the logistic function at each time point.
    """
    return m / (1 + np.exp(-k * (t - b)))

# Generate synthetic time data (years between 2000 and 2030)
time_points = np.arange(2000, 2031)

# Set parameters for the logistic function
growth_rate = 0.2
time_shift = 2024  # Set a common time shift for all technologies

# Generate synthetic adoption rate data with noise and trend (no seasonality)
np.random.seed(40)

def generate_synthetic_data(t, tech_name, starting_growth_rate):
    # Varying parameters for each technology
    max_value = np.random.uniform(0, 120)
    noise = np.random.normal(0, 1, len(t))
    trend = starting_growth_rate * (t - 2000)  # Linear trend starting from 2000

    # Generate synthetic adoption rate data using the logistic function with noise and trend
    synthetic_data = logistic_function(t, growth_rate, max_value, time_shift) + trend + noise

    # Ensure that values are within a reasonable range
    synthetic_data = np.clip(synthetic_data, 0, 100)

    # Plot the synthetic adoption rate curve
    plt.plot(t, synthetic_data, label=f'{tech_name}', linestyle='-')

    return synthetic_data

# Generate synthetic adoption rate data for each technology with different starting growth rates
aviation_synthetic = generate_synthetic_data(time_points, 'Aviation', starting_growth_rate=0.5)
ai_synthetic = generate_synthetic_data(time_points, 'AI', starting_growth_rate=0.6)
composite_synthetic = generate_synthetic_data(time_points, 'Composite', starting_growth_rate=0.18)
space_synthetic = generate_synthetic_data(time_points, 'Space', starting_growth_rate=0.4)
laser_synthetic = generate_synthetic_data(time_points, 'Laser', starting_growth_rate=1.2)

plt.xlabel("Year")
plt.ylabel("Growth Rate (%)")
plt.grid(True)
plt.title('Adoption Rate with Synthetic Data (2000-2030)', fontsize=16)
plt.legend()
plt.show()
```
